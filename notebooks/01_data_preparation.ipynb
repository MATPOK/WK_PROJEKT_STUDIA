```python
#Wybranie danych treningowych 
import os
import shutil
import pandas as pd
import matplotlib.pyplot as plt
import cv2

#KONFIGURACJA ŚCIEŻEK 
source_dir = '../data/Train' 
target_dir = '../data/selected_signs'

# Słownik z wybranymi klasami
selected_classes = {
    1: 'Ograniczenie_30',
    12: 'Droga_z_pierwszenstwem',
    13: 'Ustap_pierwszenstwa',
    14: 'Stop',
    17: 'Zakaz_wjazdu'
}

os.makedirs(target_dir, exist_ok=True)

# Kopiowanie zdjęć wybranych klas
for class_id, class_name in selected_classes.items():
    str_id_zeros = f"{class_id:05d}"
    str_id_normal = str(class_id)
    
    source_class_dir = os.path.join(source_dir, str_id_zeros)
    if not os.path.exists(source_class_dir):
        source_class_dir = os.path.join(source_dir, str_id_normal)
        
    target_class_dir = os.path.join(target_dir, class_name)
    
    os.makedirs(target_class_dir, exist_ok=True)
    
    if os.path.exists(source_class_dir):
        images = os.listdir(source_class_dir)
        images = [img for img in images if img.endswith(('.ppm', '.png', '.jpg'))]
        
        for img_name in images:
            src_path = os.path.join(source_class_dir, img_name)
            dst_path = os.path.join(target_class_dir, img_name)
            shutil.copy2(src_path, dst_path)
            
        print(f"Skopiowano {len(images)} zdjęć dla klasy: {class_name}")
    else:
        print(f"Nie znaleziono folderu źródłowego dla klasy {class_name} ({class_id})")

print("Zakończono kopiowanie zdjęć.")
```
